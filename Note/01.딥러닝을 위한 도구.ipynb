# Numpy

```python
import numpy as np
print(np.__version__)
```

### Numpy로 배열 만들기

```python
# array() 함수로 2차원 배열 만들기

my_arr = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(my_arr)
```

```python
# 넘파이 배열에서 요소 선택하기
# 30만 출력하기
my_arr[0][2]
```

```python
# 넘파이 내장 함수 사용하기
np.sum(my_arr)
```

# Matplotlib

### matplotlib으로 그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
# 선 그래프 그리기
plt.plot(
    [1,2,3,4,5],
    [1,4,9,14,25]
)
plt.show()
```

```python
# 산점도(산포도, 스캐터) 그리기

plt.scatter(
    [1,2,3,4,5],
    [1,4,9,14,25]
)
plt.show()
```

```python
# 넘파이 배열로 산점도 그리기

x = np.random.randn(1000) # 표준정규분포(평균:0, 표준편차:1)인 난수 생성 
y = np.random.randn(1000) # 표준정규분포(평균:0, 표준편차:1)인 난수 생성 

plt.scatter(
    x,
    y
)

plt.show()
```
